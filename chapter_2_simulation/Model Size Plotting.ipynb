```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_variable = [x for x in range(17)]
num_param = [4* 2**x * (2**x + 1) for x in range(17)]
```

```python
kb = 2**10
mb = 2**20
gb = 2**30
tb = 2**40
from math import log
kb_log = log(kb)
mb_log = log(mb)
gb_log = log(gb)
tb_log = log(tb)
```

```python
memory_df = pd.DataFrame(zip(num_variable, num_param), columns = ['p','bytes'])
```

```python
fig, ax = plt.subplots(1,1, figsize = (8, 10))
sns.lineplot(data = memory_df, x = "p", y = 'bytes', ax = ax)
ax.hlines(y = gb, xmin = 0, xmax = 16, linestyles = "dashed", color = 'red')
ax.annotate("1 Giga Byte (GB)", xy = (1, gb+90000000), color = 'red')

fig.savefig("memory_by_num_var.png")
```

```python
import numpy as np
```

```python
memory_df['log_bytes'] = np.log(memory_df.bytes)
```

```python
fig, ax = plt.subplots(1,1, figsize = (8, 10))
sns.lineplot(data = memory_df, x = "p", y = 'log_bytes')
ax.hlines(y = gb_log, xmin = 0, xmax = 16, linestyles = "dashed", color = 'red')
ax.hlines(y = mb_log, xmin = 0, xmax = 16, linestyles = "dashed", color = 'red')
ax.hlines(y = kb_log, xmin = 0, xmax = 16, linestyles = "dashed", color = 'red')

ax.annotate("1 Giga Byte (GB)", xy = (1, gb_log + .15), color = 'red')
ax.annotate("1 Mega Byte (MB)", xy = (1, mb_log + .15), color = 'red')
ax.annotate("1 Kilo Byte (KB)", xy = (12, kb_log + .15), color = 'red')

fig.savefig("log_memory_by_num_var.png")
```
